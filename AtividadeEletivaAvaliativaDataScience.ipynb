```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats 

grupo1 = [12, 15, 18, 22, 22, 25, 28, 30, 35, 40]
grupo2 = [15, 18, 21, 25, 25, 28, 31, 33, 38, 43]

df= pd.DataFrame({'Grupo 1': grupo1, 'Grupo2': grupo2})

# Média
media1 = np.mean(grupo1)
media2 = np.mean(grupo2)

# Variância
var1 = np.var(grupo1, ddof=1)  
var2 = np.var(grupo2, ddof=1)

# IQR
iqr1 = stats.iqr(grupo1)
iqr2 = stats.iqr(grupo2)

# Exibindo resultados
print(f"Média G1: {media1}, Variância G1: {var1}, IQR G1: {iqr1}")
print(f"Média G2: {media2}, Variância G2: {var2}, IQR G2: {iqr2}")

# Teste de normalidade
shapiro1 = stats.shapiro(grupo1)
shapiro2 = stats.shapiro(grupo2)

print("Grupo 1 - W:", shapiro1.statistic, "p-valor:", shapiro1.pvalue)
print("Grupo 2 - W:", shapiro2.statistic, "p-valor:", shapiro2.pvalue)

# Verificar se os dados são normais:
# Se p > 0.05, normal

# Teste t amostrars independentes
t_stat, p_valor = stats.ttest_ind(grupo1, grupo2)
print("Teste t: t =", t_stat, ", p-valor =", p_valor)

# Histogramas
plt.figure(figsize=(10,4))
plt.subplot(1, 2, 1)
sns.histplot(grupo1, kde=True, bins=5, color='blue')
plt.title("Histograma - Grupo 1")

plt.subplot(1, 2, 2)
sns.histplot(grupo2, kde=True, bins=5, color='green')
plt.title("Histograma - Grupo 2")
plt.tight_layout()
plt.show()

# Boxplot comparando os dois grupos
plt.figure(figsize=(6,4))
sns.boxplot(data=df)
plt.title("Boxplot - Grupo 1 vs Grupo 2")
plt.ylabel("Idade")
plt.show()


```
